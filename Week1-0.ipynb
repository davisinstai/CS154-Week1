To evaluate whether two values are equal, we can use two equals signs between them, `==`. The expression will evaluate to either `True` or `False`.

To evaluate whether two values are *not* equal, we use `!=`.

```python
# Run this code cell to determine whether the values are not equal
42 != 15
```

```python
# Run this code cell to determine whether the values are equal
42 != 42.0
```

Please note, **`==` is not the same as `=`!** Only one of those tests whether two things are equal! The other one is reserved for variable assignment, which we will learn about next time.
